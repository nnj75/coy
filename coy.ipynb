```python
from selenium import webdriver
import chromedriver_binary
import pandas as pd

driver = webdriver.Chrome()
```

```python
driver.get('http://www.espn.com/nba/history/awards/_/id/34')
```

```python
seasons = []
teams = []

for i in range(40):
    elem = driver.find_element_by_xpath('//*[@id="my-players-table"]/div[1]/div/table/tbody/tr[{}]/td[1]'.format(i+3))
    season = elem.text
    seasons.append(season)
    
for i in range(40):
    elem = driver.find_element_by_xpath('//*[@id="my-players-table"]/div[1]/div/table/tbody/tr[{}]/td[3]'.format(i+3))
    team = elem.text
    teams.append(team)
```

```python
teams
```

```python
coy_season = pd.DataFrame(seasons, columns=['season'])
coy_team = pd.DataFrame(teams, columns=['team'])
coy_df = pd.concat([coy_season, coy_team], axis=1)
coy_df
```

```python
coy_df['season'].unique()
```

```python
coy_df = coy_df.replace('', '2003')
coy_df
```
