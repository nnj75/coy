```python
from selenium import webdriver
import chromedriver_binary
import time
import pandas as pd

driver = webdriver.Chrome()
```

```python
driver.get('https://www.landofbasketball.com/yearbyyear/2020_2021_standings.htm')
```

```python
teams = []
percentages = []
standings = []
seasons = []
for i in range(40):
    
    elem = driver.find_element_by_xpath('/html/body/div[2]/div[2]/main/div[5]/div[2]/table')
    w_n = elem.find_elements_by_class_name('a-top')

    elem = driver.find_element_by_xpath('/html/body/div[2]/div[2]/main/div[5]/div[3]/table')
    e_n = elem.find_elements_by_class_name('a-top')
    
    whole_n = w_n + e_n
    
    for j in range(len(w_n)):
        elem = driver.find_element_by_xpath('/html/body/div[2]/div[2]/main/div[5]/div[2]/table/tbody/tr[{}]/td[2]'.format(j+2))
        team = elem.text
        teams.append(team)
    for j in range(len(e_n)):
        elem = driver.find_element_by_xpath('/html/body/div[2]/div[2]/main/div[5]/div[3]/table/tbody/tr[{}]/td[2]'.format(j+2))
        team = elem.text
        teams.append(team)

    for j in range(len(w_n)):
        elem = driver.find_element_by_xpath('/html/body/div[2]/div[2]/main/div[5]/div[2]/table/tbody/tr[{}]/td[5]'.format(j+2))
        percentage = float(elem.text)
        percentages.append(percentage)
    for j in range(len(e_n)):
        elem = driver.find_element_by_xpath('/html/body/div[2]/div[2]/main/div[5]/div[3]/table/tbody/tr[{}]/td[5]'.format(j+2))
        percentage = float(elem.text)
        percentages.append(percentage)

    for j in range(len(w_n)):
        standing = int(j+1)
        standings.append(standing)
    for j in range(len(e_n)):
        standing = int(j+1)
        standings.append(standing)
        
    for j in range(len(whole_n)):
        season = 2021 - i
        seasons.append(season)
        
    to_previous = driver.find_element_by_xpath('/html/body/div[2]/div[2]/main/div[2]/div[1]/div[3]/div[1]')
    to_previous.click()
    
    time.sleep(5)
```

```python
team_df = pd.DataFrame(teams, columns=['team'])
standing_df = pd.DataFrame(standings, columns=['standing'])
percentage_df = pd.DataFrame(percentages, columns=['percentage'])
season_df = pd.DataFrame(seasons, columns=['season'])
```

```python
records_df = pd.concat([team_df, standing_df, percentage_df, season_df], axis=1)
```
