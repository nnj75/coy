```python
from selenium import webdriver
import chromedriver_binary
import pandas as pd
import time

driver = webdriver.Chrome()
```

```python
driver.get('http://www.espn.com/nba/history/awards/_/id/40')
```

```python
n = driver.find_elements_by_class_name('oddrow')

seasons = []
teams = []

for i in range(len(n)):
    elem = driver.find_element_by_xpath('//*[@id="my-players-table"]/div[1]/div/table/tbody/tr[{}]/td[1]'.format(i+3))
    season = elem.text
    seasons.append(season)
    
for i in range(len(n)):
    elem = driver.find_element_by_xpath('//*[@id="my-players-table"]/div[1]/div/table/tbody/tr[{}]/td[4]'.format(i+3))
    team = elem.text
    teams.append(team)
```

```python
smoy_season = pd.DataFrame(seasons, columns=['season'])
```

```python
smoy_team = pd.DataFrame(teams, columns=['team'])
```

```python
smoy_df = pd.concat([smoy_season, smoy_team], axis=1)
smoy_df
```

```python
smoy_df.to_csv('smoy.csv')
```
